```python
# -*- coding: utf-8 -*-
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 1, 100)#在0-10之间取1000个点
y = (-x*np.log2(x)-(1-x)*np.log2(1-x))/2
z = 2*x*(1-x)
# q = x*(1-x)
# x>0 1-x
# x<0 x
plt.figure(figsize=(8,4))
plt.plot(x,y,label="$entropy$",color="red",linewidth=2)
plt.plot(x,z,"b--",label="$gini$")
# plt.plot(x,q,"g--",label="$error$")

plt.xlabel("Time(s)")
plt.ylabel("Volt")
plt.title("PyPlot First Example")
plt.ylim(0,0.5)
plt.legend()
plt.show()
```
